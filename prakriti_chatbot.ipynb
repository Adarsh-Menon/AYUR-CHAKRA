```python
import transformers
from transformer_embeddings import TransformerEmbeddings
model_id = 'sentence-transformers/all-MiniLM-L6-v2'
hugface_id = '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'
transformer = TransformerEmbeddings(model_id)
```

```python
import requests
api_url = f"https://api-inference.huggingface.co/pipeline/feature-extraction/{model_id}"
headers={'organization':f"bearer{hugface_id}"}
```

```python
def query(cleaned_answers):
  response = requests.post(api_url, headers=headers, json={'inputs':cleaned_answers})
  return response.json()
```

```python
output = query(cleaned_answers)
```

```python
embeddings=pd.DataFrame(output)
```

```python
import torch
import numpy as np
dataset_embeddings = torch.from_numpy(embeddings.to_numpy()).to(torch.float)
```

```python
q=['why do i have brown hair?']
output=query(q)
query_embeddings=torch.FloatTensor(output)
```

```python
from sentence_transformers.util import semantic_search

top_answers=semantic_search(query_embeddings,dataset_embeddings,top_k=5)
```

```python
print(cleaned_answers[top_answers[0][0]['corpus_id']])
```
